```python
#nhập ma trận A cỡ 5x5 bất kì
import numpy as np
A = np.random.randint(1, 10, (5,5))
#biến A thành floating point
A = A.astype(np.float64)
for i in range(5):
    for j in range(5):
        if i > j:
            A[i][j] = A[j][i]
print(A)
```

```python
#rank của ma trận A
print(np.linalg.matrix_rank(A))
```

```python
#tính QR factorization của ma trận A
from qr_by_given_rotations import QRByGivenRotation


qr = QRByGivenRotation(A)
R, rotations = qr.decomposite()
print('Ma trận R:', R)
```

```python
def checkHessenberg(A_matrix, eps=1e-10):
    #kiem tra duoi duong cheo phu
    for i in range(0, A_matrix.shape[1]):
        for j in range(i+1, A_matrix.shape[0]):
            if abs(A_matrix[j,i]) > eps:
                print('Duoi duong cheo phu khac 0',j,i, A_matrix[j,i])
                return False
           
            
    return True
```

```python

list_QR_by_given_rotation = [qr]
counter = 0
while True:
    current_qr = list_QR_by_given_rotation[-1]
    current_qr.decomposite()
    next_A = current_qr.R.copy()
    for i in range(next_A.shape[0]):
        next_A[i] = current_qr.right_multiply_Q(next_A[i])
    list_QR_by_given_rotation.append(QRByGivenRotation(next_A))
    print(next_A)
    counter += 1
    # if counter ==30:
    #     break
    if(checkHessenberg(next_A)):
        break
print('Số lần lặp:', counter)
print('Ma trận A cuối cùng:', list_QR_by_given_rotation[-1].A)
```

```python
last_A = list_QR_by_given_rotation[-1].A
Q = np.identity(A.shape[0])
for i in range(len(list_QR_by_given_rotation)):
    for j in range(A.shape[0]):
        Q[j] =list_QR_by_given_rotation[i].right_multiply_Q(Q[j])
print('Ma trận Q:', Q)
```

```python
A_test = Q @ last_A @ Q.T
print('Ma trận A_test:', A_test, A)
```

```python
#dung scipy để tính eigenvalue và eigenvector
from scipy.linalg import eig
eigenvalues, eigenvectors = eig(A)

#tinh Av - lambda*v
for i in range(len(eigenvalues)):
    print(np.linalg.norm( A @ eigenvectors[:,i] - eigenvalues[i] * eigenvectors[:,i]))
    print('\n')
```

```python
#tinh Av - lambda*v cua minh
for i in range(len(eigenvalues)):
    print(np.linalg.norm( A @ Q[:,i] - last_A[i,i] * Q[:,i]))
    print('\n')
```
