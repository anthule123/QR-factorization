```python
#nhập ma trận A cỡ 5x5 bất kì
import numpy as np
A = np.random.randint(1, 10, (5,5))
#biến A thành floating point
A = A.astype(float)
print(A)
```

```python
#rank của ma trận A
print(np.linalg.matrix_rank(A))
```

```python
#tính QR factorization của ma trận A
from qr_by_given_rotations import QRByGivenRotation


qr = QRByGivenRotation(A)
R, rotations = qr.decomposite()
print('Ma trận R:', R)
```

```python
#Tinh A1 = RQ
A1 = R.copy()
for i in range(4):
    A1[i] = qr.right_multiply_Q(A1[i])
print('Ma trận A1:', A1)
```

```python
def checkHessenberg(A_matrix, eps=0.1):
    #kiem tra duoi duong cheo phu
    for i in range(0, A_matrix.shape[1]):
        for j in range(i+2, A_matrix.shape[0]):
            if abs(A_matrix[j,i]) > eps:
                print('Duoi duong cheo phu khac 0',j,i, A_matrix[j,i])
                return False
            
    return True
```

```python
qr2 = QRByGivenRotation(A1)
R2, rotations2 = qr2.decomposite()
print('Ma trận R2:', R2)

A2 = R2.copy()
for i in range(4):
    A2[i] = qr2.right_multiply_Q(A2[i])
print('Ma trận A2:', A2)
```

```python

list_QR_by_given_rotation = [qr]
counter = 0
while True:
    current_qr = list_QR_by_given_rotation[-1]
    current_qr.decomposite()
    next_A = current_qr.R.copy()
    for i in range(next_A.shape[0]):
        next_A[i] = current_qr.right_multiply_Q(next_A[i])
    list_QR_by_given_rotation.append(QRByGivenRotation(next_A))
    print(next_A)
    counter += 1
    if counter ==30:
        break
    if(checkHessenberg(next_A)):
        break
print('Số lần lặp:', counter)
print('Ma trận A cuối cùng:', list_QR_by_given_rotation[-1].A)
```
